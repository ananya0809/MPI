#### Serial

```python
import matplotlib.pyplot as plt
import numpy
plate = numpy.fromfile("plate.out", dtype=float).reshape((1002,1002))
plt.title("Temperature Distribution Laplace Serial (Steady State)")
plt.xlabel("Columns")
plt.ylabel("Rows")
plt.imshow(plate)
plt.colorbar(label="Temperature")
plt.imshow(plate, norm=matplotlib.colors.LogNorm(0.1,50,clip=True))
```
